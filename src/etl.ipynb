```python
import pandas as pd
```

```python

#Extract data from csv file
df = pd.read_csv("../data/raw_data/orders.csv")
df.head()
```

```python
# Transform data from csv file
df['order_date'] = pd.to_datetime(df['order_date'])
df['total_amount'] = df['price'] * df['quantity']
df = df.drop_duplicates()
df.head()
```

```python
# Write cleaned orders to CSV
df.to_csv("../data/processed_data/orders_clean.csv", index=False)
```
